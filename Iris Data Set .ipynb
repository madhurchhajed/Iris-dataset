```python
import pandas as pd
import matplotlib.pyplot as plt
filepath = "iris.csv"
data = pd.read_csv(filepath)
print(data.iloc[:5])
```

```python
data['sepal_area']=data.sepal_length * data.sepal_width
print(data.iloc[:5,-3:])
```

```python
print(data.mean())

print(data.median())

print(data.mode())
```

```python
print(data.describe())
```

```python
sample = (data.sample(n=5,replace=False,random_state=42))
print(sample.iloc[:,-3:])
```

```python
data['abbrev'] = (data.class.apply(lambda x:x.replace('Iris-',' ')))
```

```python
group_sizes = (data.groupby('class').size())
group_sizes
```

```python
plt.plot(data.sepal_length,data.sepal_width,ls='',marker='o',color='red')
```

```python
plt.plot(data.sepal_length,data.sepal_width,ls='',marker='o',label='sepal')
plt.plot(data.petal_length,data.petal_width,ls='',marker='o',label='petal')
```
